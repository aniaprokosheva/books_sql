## Описание данных

## 1 Сколько книг вышло после 1 января 2000 года

```python
query1 = ''' 
SELECT COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01'
'''

request1 = pd.io.sql.read_sql(query1, con = engine)


print(f'После 1 января 2000 года вышло {request1.iloc[0,0]} книг.')
```

## 2 Количество обзоров и средняя оценка для каждой книги

```python
query2 = ''' 

SELECT 
    books.title AS book_name,
    books.book_id AS id,
    rev.n_rev AS n_reviews,
    rat.avg_rat AS avg_rating
FROM books LEFT JOIN 
                (SELECT
                    book_id,
                    AVG(rating) AS avg_rat
                FROM
                    ratings
                GROUP BY 
                    book_id
                ) AS rat ON rat.book_id = books.book_id
        LEFT JOIN
                (SELECT 
                    book_id,
                    COUNT(review_id) AS n_rev
                FROM
                    reviews
                GROUP BY 
                    book_id
                ) AS rev ON rev.book_id = books.book_id 
''' 

request2 = pd.io.sql.read_sql(query2, con = engine)
print('Количество обзоров(n_reviews) и средняя оценка(avg_rating) для каждой книги:')
request2
```

## 3 Издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query3 = ''' 
SELECT
    P.publisher_id,
    P.publisher,
    P.n_books
FROM
    (SELECT 
        books.publisher_id AS publisher_id,
        COUNT(books.book_id) AS n_books,
        publishers.publisher AS publisher 
    FROM 
        books 
        LEFT JOIN publishers ON books.publisher_id=publishers.publisher_id

    WHERE 
        num_pages>50
    GROUP BY
        books.publisher_id,
        publishers.publisher
    ORDER BY 
        n_books DESC) AS P
LIMIT 1 
''' 

request3 = pd.io.sql.read_sql(query3, con = engine)
print('Издательство, которое выпустило наибольшее число книг:')
request3
```

## 4 - Автор с самой высокой средней оценкой книг

Учитываются книги с 50 и более оценками

```python
query4 = ''' 
SELECT 
    aut_avg.author,
    AVG(aut_avg.avg_rating) AS avg_avg_rating
FROM
        (SELECT 
            authors.author AS author, 
            AVG(ratings.rating) AS avg_rating
        FROM books
            LEFT JOIN authors ON books.author_id = authors.author_id
            LEFT JOIN ratings ON books.book_id = ratings.book_id
        GROUP BY 
             books.book_id, 
             authors.author_id
        HAVING
            COUNT(ratings.rating_id) >=50 
        ) AS aut_avg 
GROUP BY    
    aut_avg.author
ORDER BY
    avg_avg_rating DESC
LIMIT 1
 ''' 

request4 = pd.io.sql.read_sql(query4, con = engine)
print('Автор с самой высокой средней оценкой книг:')
request4
```

## 5 - Среднее количество обзоров от пользователей

Учитываются пользователи, которые поставили больше 48 оценок

```python
query5 = ''' 

SELECT
    AVG(rev.n_reviews)
FROM
    (SELECT
        username,
        COUNT(review_id) AS n_reviews
    FROM
        reviews
    WHERE 
        username IN (
                SELECT 
                    username
                FROM 
                    ratings
                GROUP BY 
                    username
                HAVING 
                    COUNT(rating_id) > 48
                    )
    GROUP BY 
        username
    ) AS rev
''' 

request5 = pd.io.sql.read_sql(query5, con = engine)
print(f'Среднее количество обзоров от пользователей {request5.iloc[0,0]}.')
```
